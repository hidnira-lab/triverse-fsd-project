### 3.1 Pemeriksaan Outlier (Metode IQR)

Khusus untuk 7 fitur kebiasaan numerik yang akan dipakai untuk clustering, dicek outlier-nya dengan metode IQR: nilai dianggap outlier kalau berada di luar rentang `Q1 - 1.5*IQR` sampai `Q3 + 1.5*IQR`.

```python
habit_numeric_cols_check = [
    "study_hours_per_day", "social_media_hours", "netflix_hours",
    "attendance_percentage", "sleep_hours", "exercise_frequency", "mental_health_rating"
]

outlier_rows = []
for col in habit_numeric_cols_check:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    n_outlier = ((df[col] < lower_bound) | (df[col] > upper_bound)).sum()
    outlier_rows.append({
        "fitur": col,
        "lower_bound": round(lower_bound, 2),
        "upper_bound": round(upper_bound, 2),
        "jumlah_outlier": n_outlier,
        "persentase": round(n_outlier / len(df) * 100, 2)
    })

outlier_df = pd.DataFrame(outlier_rows).sort_values("jumlah_outlier", ascending=False)
outlier_df
```

Dari 7 fitur kebiasaan numerik, ditemukan outlier tipis di 5 fitur: `study_hours_per_day` (7 baris), `social_media_hours` (5 baris), `netflix_hours` (4 baris), `attendance_percentage` (3 baris), dan `sleep_hours` (2 baris). Total ada 21 baris (2,1%) yang punya outlier di setidaknya satu fitur; `exercise_frequency` dan `mental_health_rating` tidak punya outlier sama sekali.

Berbeda dengan Studi Kasus 1, fitur di sini kontinu (bukan skala Likert terbatas), jadi outlier yang terdeteksi bukan cuma nilai batas skala. Meski begitu, nilai-nilainya tetap masuk akal secara perilaku: misalnya `study_hours_per_day` maksimum 8,3 jam/hari (mahasiswa yang sangat rajin) dan `attendance_percentage` minimum 56% (mahasiswa dengan kehadiran rendah) sama-sama kondisi nyata yang bisa terjadi, bukan kesalahan input data.

Keputusan: outlier dipertahankan, tidak dihapus atau di-cap. Jumlahnya sangat kecil (2,1% dari total data) dan nilai-nilainya adalah representasi valid dari variasi kebiasaan mahasiswa, yang justru relevan untuk tujuan clustering, yaitu menemukan kelompok kebiasaan yang berbeda. Menghapus mahasiswa dengan kebiasaan ekstrem justru mengurangi variasi yang ingin ditangkap oleh clustering.

## 7. Preprocessing

```python
df_features = df.copy()

diet_map = {"Poor": 0, "Fair": 1, "Good": 2}
binary_map = {"No": 0, "Yes": 1}

df_features["diet_quality_encoded"] = df_features["diet_quality"].map(diet_map)
df_features["part_time_job_encoded"] = df_features["part_time_job"].map(binary_map)
df_features["extracurricular_participation_encoded"] = df_features["extracurricular_participation"].map(binary_map)

clustering_features = habit_numeric_cols + [
    "diet_quality_encoded", "part_time_job_encoded", "extracurricular_participation_encoded"
]

X = df_features[clustering_features]
print("Jumlah fitur clustering:", X.shape[1])
print("Missing value pada feature matrix:", X.isnull().sum().sum())
X.head()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled_df = pd.DataFrame(X_scaled, columns=clustering_features)
print("Shape feature matrix ter-scaling:", X_scaled_df.shape)
X_scaled_df.describe().T[["mean", "std"]]
```

Encoding: `diet_quality` di-encode ordinal (Poor=0, Fair=1, Good=2) karena punya urutan tingkatan yang jelas; `part_time_job` dan `extracurricular_participation` di-encode biner (No=0, Yes=1). Semua fitur clustering ini numerik dan bebas dari missing value (0 missing di feature matrix), sehingga sudah bisa langsung di-scaling. Hasil `StandardScaler` menunjukkan setiap fitur punya mean ~0 dan std ~1, siap dipakai untuk clustering.

## 8. Training K-Means dan Pemilihan Jumlah Cluster

Bagian ini mencari jumlah cluster yang paling masuk akal dengan **Elbow Method**, lalu memvalidasinya menggunakan **Silhouette Score**. Setelah `k` terpilih, model K-Means dilatih pada feature matrix yang sudah di-scaling.

### Kenapa K-Means, Bukan Hierarchical Clustering atau DBSCAN

Ada tiga pertimbangan utama sebelum memilih algoritma clustering untuk kasus ini:

1. **Ukuran dataset.** Hierarchical Clustering (agglomerative) punya kompleksitas komputasi dan memori sekitar O(n²) sampai O(n³) karena harus menghitung dan menyimpan matriks jarak antar semua pasangan data. Dengan 1.000 baris, ini sudah mulai berat dibanding K-Means yang kompleksitasnya jauh lebih ringan (linear terhadap jumlah data dan jumlah cluster per iterasi).
2. **Tidak ada kebutuhan struktur hierarki.** Tujuan studi kasus ini adalah mengelompokkan mahasiswa ke sejumlah kelompok kebiasaan yang jelas untuk dianalisis dan dilaporkan, bukan membangun dendrogram atau melihat hubungan bertingkat antar sub-kelompok. Hierarchical Clustering lebih cocok kalau memang butuh eksplorasi struktur bertingkat itu, yang bukan kebutuhan di sini.
3. **Tidak ada indikasi pola density yang tidak beraturan.** DBSCAN unggul untuk data dengan bentuk cluster sembarang dan kepadatan yang bervariasi, serta cocok kalau ada noise yang perlu diabaikan otomatis. Dari EDA, fitur-fitur kebiasaan di dataset ini terdistribusi mendekati normal dan kontinu tanpa tanda density yang tidak beraturan, dan pemeriksaan outlier sebelumnya juga menunjukkan datanya bersih (outlier di bawah 3% dan tetap masuk akal). Tidak ada alasan kuat memakai DBSCAN di sini, apalagi DBSCAN tidak memberi jumlah cluster secara eksplisit, padahal jumlah cluster yang jelas penting untuk diinterpretasikan di laporan.

K-Means juga bisa dipasangkan langsung dengan Elbow Method dan Silhouette Score untuk memilih jumlah cluster secara terukur, dan hasilnya (nilai `k`) mudah dijelaskan ke pembaca laporan dibanding hasil DBSCAN yang jumlah cluster-nya muncul otomatis dari parameter epsilon dan minPts.

Sebelum masuk ke kode, alur training dibagi jadi tiga tahap. Pertama, menyiapkan library yang dibutuhkan untuk K-Means, PCA, dan silhouette. Kedua, mencoba beberapa nilai `k` untuk melihat titik elbow sekaligus skor silhouette terbaik. Ketiga, melatih model final, memvisualisasikan hasil cluster, lalu membaca karakter tiap cluster dan membandingkan `exam_score`.

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score

print("K-Means, PCA, dan silhouette siap dipakai.")
```

Sel pertama hanya menyiapkan komponen analisis. `KMeans` dipakai untuk membentuk cluster, `PCA` untuk menurunkan dimensi ke 2D agar bisa divisualisasikan, dan `silhouette_score` untuk mengevaluasi seberapa baik pemisahan cluster yang dihasilkan.

```python
k_values = range(2, 11)
inertias = []
silhouette_scores = []

for k in k_values:
    model = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = model.fit_predict(X_scaled)
    inertias.append(model.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, labels))

metrics_df = pd.DataFrame({
    "k": list(k_values),
    "inertia": inertias,
    "silhouette_score": silhouette_scores,
})

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].plot(metrics_df["k"], metrics_df["inertia"], marker="o")
axes[0].set_title("Elbow Method")
axes[0].set_xlabel("Jumlah Cluster (k)")
axes[0].set_ylabel("Inertia")
axes[0].grid(True, alpha=0.3)

axes[1].plot(metrics_df["k"], metrics_df["silhouette_score"], marker="o", color="tab:green")
axes[1].set_title("Silhouette Score")
axes[1].set_xlabel("Jumlah Cluster (k)")
axes[1].set_ylabel("Silhouette Score")
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(metrics_df.round(4).to_string(index=False))
```

Di tahap ini, beberapa nilai `k` diuji dari 2 sampai 10. Untuk setiap `k`, model K-Means dihitung inertia-nya untuk Elbow Method dan juga silhouette score-nya untuk melihat kualitas pemisahan cluster. Grafik yang muncul membantu memilih `k` yang seimbang antara kompaksi cluster dan pemisahan antarcluster.

```python
k_selected = metrics_df.loc[metrics_df["silhouette_score"].idxmax(), "k"]

kmeans = KMeans(n_clusters=int(k_selected), random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(X_scaled)

clustered_df = df_features.copy()
clustered_df["cluster"] = cluster_labels

silhouette_final = silhouette_score(X_scaled, cluster_labels)
print(f"k terpilih: {k_selected}")
print(f"Silhouette Score final: {silhouette_final:.4f}")

pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)
pca_df = pd.DataFrame(X_pca, columns=["PC1", "PC2"])
pca_df["cluster"] = cluster_labels

plt.figure(figsize=(10, 7))
sns.scatterplot(data=pca_df, x="PC1", y="PC2", hue="cluster", palette="tab10", s=60, alpha=0.85)
plt.title("Visualisasi Cluster K-Means dengan PCA")
plt.legend(title="Cluster")
plt.tight_layout()
plt.show()

cluster_profile = clustered_df.groupby("cluster")[clustering_features + ["exam_score"]].mean().round(2)
print("Rata-rata fitur per cluster:")
print(cluster_profile.to_string())

overall_means = clustered_df[clustering_features + ["exam_score"]].mean()
relative_profile = (clustered_df.groupby("cluster")[clustering_features].mean() - overall_means[clustering_features]).round(2)
print("\nDeviasi mean cluster dari rata-rata keseluruhan:")
print(relative_profile.to_string())

print("\nRata-rata exam_score per cluster:")
print(clustered_df.groupby("cluster")["exam_score"].mean().round(2).to_string())
```

Setelah `k` terbaik ditemukan, model final dilatih ulang menggunakan seluruh feature matrix yang sudah di-scaling. Hasil label cluster lalu digabung kembali ke dataframe asli supaya profil tiap cluster bisa dihitung. PCA dipakai untuk memproyeksikan 10 fitur ke 2 dimensi, sehingga hasil cluster lebih mudah dibaca lewat scatter plot. Bagian akhir merangkum rata-rata fitur per cluster, deviasi terhadap rata-rata keseluruhan, dan rata-rata `exam_score` sebagai insight tambahan.

### Interpretasi Karakteristik Cluster

Silhouette Score final (0.1444) tergolong rendah, artinya kedua cluster saling overlap dan tidak terpisah tajam. Ini wajar untuk data kebiasaan mahasiswa yang variasinya cenderung kontinu, bukan diskret, dan visualisasi PCA 2D juga menunjukkan kedua cluster memang berdekatan.

Pembeda paling tajam antar cluster ternyata bukan pola "rajin vs malas belajar" secara umum seperti dugaan awal di bagian EDA, melainkan status pekerjaan paruh waktu:

- Cluster 0: seluruh anggotanya (100%) memiliki `part_time_job`. Fitur kebiasaan lain (jam belajar, media sosial, screen time, tidur, olahraga) hanya berbeda tipis dari rata-rata keseluruhan.
- Cluster 1: seluruh anggotanya (0%) tidak memiliki `part_time_job`.

Fitur kebiasaan lain punya deviasi mendekati nol terhadap rata-rata keseluruhan di kedua cluster. Ini yang menjelaskan kenapa Silhouette Score-nya rendah: satu-satunya sinyal pemisah yang kuat cuma `part_time_job`, sementara 9 fitur lainnya nyaris tidak berkontribusi membedakan cluster.

Dari sisi akademik, Cluster 0 (mahasiswa dengan part-time job) punya `attendance_percentage` dan rata-rata `exam_score` sedikit lebih rendah dibanding Cluster 1, meski selisihnya kecil (sekitar 1 poin). Insight ini relevan untuk kebijakan dukungan akademik bagi mahasiswa yang bekerja sambil kuliah.
